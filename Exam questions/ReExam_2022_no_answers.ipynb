1. The chromatic number of a graph is the minimum number of colors needed to color the vertices of the graph such that no two adjacent vertices have the same color. In other words, it is the minimum number of colors required to assign a unique color to each vertex in such a way that no two adjacent vertices have the same color.Non-trivial example: Consider a graph with five vertices arranged in the shape of a pentagon (a five-sided polygon) where each vertex is connected to every other vertex. This graph is known as a complete graph with five vertices, denoted as K5. The chromatic number of this graph is 5 because each vertex is adjacent to every other vertex, and thus, a minimum of 5 colors is required to color the vertices such that no two adjacent vertices have the same color.

2. A bipartite graph is a graph whose vertices can be divided into two disjoint sets such that there are no edges connecting vertices within the same set. In other words, it is a graph that can be separated into two independent sets of vertices, and there are no edges between vertices within each set. Non-trivial example: Consider a graph with six vertices, divided into two sets of three vertices each. Let's denote the two sets as A and B. In this graph, all the vertices in set A are connected to all the vertices in set B, but there are no connections between vertices within the same set. This graph is an example of a bipartite graph.

3. The chromatic number of a bipartite graph is always 2. Since a bipartite graph can be divided into two disjoint sets of vertices, there are no edges between vertices within each set. Therefore, it is always possible to color the vertices of a bipartite graph using only two colors, one for each set, without any conflicts.

4. The chromatic number of a complete graph with n vertices, denoted as Kn, is always n. In a complete graph, every vertex is connected to every other vertex, resulting in maximum adjacency. Therefore, to color a complete graph, a minimum of n colors is required, where each color is assigned to a different vertex.

5. The chromatic number of a cycle graph with n vertices, denoted as Cn, is 2 if n is even, and 3 if n is odd. In a cycle graph, each vertex is connected to its adjacent vertices, forming a closed loop. If the number of vertices is even, the graph can be divided into two sets of vertices, where each set alternates in the cycle. Therefore, only two colors are needed to color the vertices. If the number of vertices is odd, there is an additional vertex that cannot be connected to both of its adjacent vertices without creating an edge between them. In this case, three colors are required to avoid adjacent vertices having the same color.

6. If a graph contains a clique of size n, where a clique is a subset of vertices where each vertex is connected to every other vertex in the subset, then the chromatic number of the graph is also n. Since all the vertices in the clique are adjacent to each other, each vertex in the clique must have a different color. Thus, a minimum of n colors is needed to color the graph without any conflicts.

From Wikipedia[[1]]: "a maximum common induced subgraph of two graphs G and H is a graph that is an induced subgraph of both G and H, and that has as many vertices as possible".

Given two simple, undirected graphs $G$ and $H$ one way of finding the maximum common induced subgraph is by first constructing the *modular product graph*[[2]] $M$, and then finding all cliques in $M$. If $G$ has nodes $a, b, c, d, ...$, and H has nodes $1, 2, 3, 4, ...$ then $M$ will have as nodes all possible combinations: $V(M) = (a, 1), (a, 2), ..., (a, |H|), (b, 1), ..., (b, |H|), ..., (|G|, 1), ..., (|G|, |H|)$. $M$ will contain an edge if the edge is either in both graphs, or in neither. In other words:
Edge $((u1, v1), (u2, v2)) \in E(M)$ if either $(u1, u2) \in E(G)$ *and* $(v1, v2) \in E(H)$; or $(u1, u2) \notin E(G)$ *and* $(v1, v2) \notin E(H)$. 

For example, if $(a, b) \in E(G)$ and $(2, 6) \in E(H)$, then $((a, 2), (b, 6)) \in E(M)$. Or, if $(a, c) \notin E(G)$ and $(3, 6) \notin E(H)$ then $((a, 3), (c, 6)) \in E(M)$.

Clique(s) of $M$ will consist of nodes in $M$, which in turn correspond to nodes in $G$ and $H$. For example, if $M$ contains a clique ${(a, 3), (b, 2), (c, 1)}$ it means the induced subgraph ${a, b, c} \subseteq V(G)$ is isomorphic to the induced subgraph ${1, 2, 3} \subseteq V(H)$. It is often convenient to express this as a mapping of nodes between $G$ and $H$. In this example, the Python dictionary `{'a': 3, 'b': 2, 'c': 1}`.
