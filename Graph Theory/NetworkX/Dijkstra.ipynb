```python
# My Dijkstra's Algorithm
seed(1)
h = nx.barabasi_albert_graph(10,2)

for e in h.edges():
    h.edges[e]['weight'] = randint(1,5)

print(nx.get_edge_attributes(h, 'weight'))    

pos = nx.spring_layout(h, weight='w')
nx.draw_networkx(h, pos=pos)
nx.draw_networkx_edge_labels(h, pos=pos, edge_labels=nx.get_edge_attributes(h, 'weight'));
```

```python

def dijkstra(h, start):
    visited = []
    unvisited = list(h.nodes)
    dijkstra_dict = {}
    current_vertex = start
    h1 = h.copy()

    # make a dictionary to store dijkstra factors
    for node in h1.nodes:

        if node == start:
            dijkstra_dict[node]= [0, ' ']
   
        else:
            dijkstra_dict[node] = [float('inf'), ' ']

    # As long as unvisited nodes' list is not empty irritate
    while len(unvisited) != 0: 
        assistant = 0
        assistant_vertex = ''

        # for each neibour for the current vertex find its weight  
        for (u, v) in h1.edges(current_vertex):
            weight = h1.edges[(u, v)]['weight'] + dijkstra_dict[u][0] 

            # if the weight is smaller than the privous one for a node, replace it 
            if weight < dijkstra_dict[v][0]:
                dijkstra_dict[v][0] = weight
                dijkstra_dict[v][1] = u

        # add the current node to visited list remove from unvisited
        visited.append(current_vertex)
        unvisited.remove(current_vertex) 

        # find the next current node
        for node in unvisited:
            if assistant == 0 or assistant > dijkstra_dict[node][0]:
                assistant = dijkstra_dict[node][0]
                assistant_vertex = node

        # correct the graph and assign the next node
        h1.remove_node(current_vertex) 
        current_vertex = assistant_vertex
    return dijkstra_dict

print(dijkstra(g,'a'))
```

```python

def dijkstra_path(start_node, end_node,g):

    # make dijkstra dictionary base on the starting point and graph
    dijkstra_dict = dijkstra(g,start_node)
    dijkstra_path = [end_node]
    current_node = end_node
    path_lenght = dijkstra_dict[end_node][0]

    # find the shortest path
    while current_node != start_node:
        dijkstra_path.append(dijkstra_dict[current_node][1])
        current_node = dijkstra_dict[current_node][1]

    return dijkstra_path, path_lenght

print(dijkstra_path('a', 'f', g))
```
